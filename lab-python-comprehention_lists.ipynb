```python
def initialize_inventory(products):
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory

def get_customer_orders():
    num_orders = int(input("Enter the number of customer orders: "))
    customer_orders = [input("Enter the name of a product that a customer wants to order: ") for _ in range(num_orders)]
    return customer_orders

def calculate_total_price(customer_orders):
    prices = []
    for product in customer_orders:
        while True:
            try:
                price = float(input(f"Enter the price of {product}: "))
                prices.append(price)
                break
            except ValueError:
                print("Invalid input. Please enter a valid price.")
    total_price = sum(prices)
    return total_price

def update_inventory(inventory, customer_orders):
    for order in customer_orders:
        if order in inventory:
            inventory[order] -= 1
    inventory = {product: quantity for product, quantity in inventory.items() if quantity > 0}
    return inventory

def run_order_management():
    # Initial Inventory
    products = ["t-shirt", "mug", "hat", "book", "keychain"]
    
    # Step 1: Initialize Inventory
    inventory = initialize_inventory(products)
    
    # Step 2: Get Customer Orders
    customer_orders = get_customer_orders()
    
    # Filter orders to only include products in the inventory
    valid_orders = [order for order in customer_orders if order in inventory]
    
    # Step 3: Update Inventory
    inventory = update_inventory(inventory, valid_orders)
    
    # Step 4: Calculate Total Price
    total_price = calculate_total_price(valid_orders)
    
    # Print Customer Orders
    print("\nCustomer Orders:")
    for order in valid_orders:
        print(order)
    
    # Print Order Statistics
    print("\nOrder Statistics:")
    total_products_ordered = len(valid_orders)
    percentage_unique_products_ordered = (len(set(valid_orders)) / len(products)) * 100
    print(f"Total Products Ordered: {total_products_ordered}")
    print(f"Percentage of Unique Products Ordered: {percentage_unique_products_ordered:.1f}%")
    
    # Print Updated Inventory
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
    
    # Print Total Price
    print(f"Total Price: {total_price}")

if __name__ == "__main__":
    run_order_management()
```
